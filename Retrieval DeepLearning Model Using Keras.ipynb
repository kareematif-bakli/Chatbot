```python
import nltk
from nltk.stem.lancaster import LancasterStemmer
stemmer = LancasterStemmer()
from nltk.corpus import stopwords
import numpy
import tensorflow
import random
import json
import pickle
import pandas as pd

words = []
labels = []
docs_x = []
docs_y = []
stop_words= set(stopwords.words("english"))
filered_sentence = []

#nltk.download('punkt')

data = pd.read_csv("Data.csv")
```

```python
for pattern in data["Questions"]:
    wrds = nltk.word_tokenize(pattern)
    words.extend(wrds)
    docs_x.append(wrds)
for pattern in data["Tags"]:    
    docs_y.append(pattern)
    if pattern not in labels:
        labels.append(pattern)

words = [stemmer.stem(w.lower()) for w in words if w != "?"]

words = sorted(list(set(words)))

labels = sorted(labels)

training = []
output = []

out_empty = [0 for _ in range(len(labels))]

print(words)
```

```python
for x, doc in enumerate(docs_x):
    bag = []

    wrds = [stemmer.stem(w.lower()) for w in doc]
    
    print(wrds)
    print("***************")
    for w in words:
        print(w)
        if w in wrds:
            bag.append(1)
        else:
            bag.append(0)
    
    output_row = out_empty[:]
    output_row[labels.index(docs_y[x])] = 1

    training.append(bag)
    output.append(output_row)

training = numpy.array(training)
output = numpy.array(output)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.layers import Dense
model = Sequential()
#model.Input=[None,len(training[0])]

model.add(Dense(units=8,input_dim=len(training[0])))
model.add(Dense(units=8))
model.add(Dense(units=8))
model.add(Dense(len(output[0]), activation='softmax'))




model.compile(loss="categorical_crossentropy",
              optimizer="sgd",metrics=['accuracy']

print(training.shape)
print(output.shape)

model.fit(training,output,epochs=1000,batch_size=20)
#score = model.evaluate(x_test, y_test, batch_size=128)


#model.train_on_batch(x_batch, y_batch)

model.summary()
```

```python
def bag_of_words(s,words):
    bag = [0 for _ in range(len(words))]

    s_words = nltk.word_tokenize(s)
    s_words = [stemmer.stem(word.lower()) for word in s_words]

    for se in s_words:
        for i, w in enumerate(words):
            if w == se:
                bag[i] = 1
                
    #bag=numpy.array(bag)
   # bag.reshape(386,)
   # print(bag.shape)
    return bag
```
